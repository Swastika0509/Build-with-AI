```python
from sklearn.ensemble import RandomForestClassifier
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('ToyotaCorolla.csv')
```

```python
plt.figure(figsize=(14, 4))
sns.violinplot(x='Fuel_Type', y='Price', data=data, hue='Fuel_Type', palette='bright', legend=False)
plt.title("Price Distribution by Fuel Type")
plt.show()
```

```python
plt.figure(figsize=(14, 4))
sns.barplot(x='Fuel_Type', y='Price', data=data, hue='Fuel_Type', palette='muted', legend=False)
plt.title("Fuel Type vs Price")
plt.show()
```

```python
plt.figure(figsize=(14, 4))
sns.countplot(x='Fuel_Type', data=data, hue='Fuel_Type', palette='muted', legend=False)
plt.title("Count of Cars by Fuel Type")
plt.show()
```

```python
numeric_data = data.select_dtypes(include=['number'])

plt.figure(figsize=(30,10))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
X = data[['Price', 'Age_08_04', 'HP','Price','Doors','Cylinders','Gears','Airbag_1','Airbag_2']]
sns.histplot(data['Price'], bins=30)
plt.title("Price Distribution")
```

```python
X = data[['Age_08_04', 'KM', 'HP', 'Met_Color', 'Automatic', 'cc', 'Doors', 'Cylinders', 'Gears', 'Weight', 'Guarantee_Period', 'Powered_Windows', 'Power_Steering']]
y = data['Price']
model = RandomForestClassifier()
model.fit(X, y)
```

```python
importances = model.feature_importances_
features = X.columns
```

```python
plt.figure(figsize=(10,6))
plt.xticks(rotation=45)
colors = sns.color_palette("bright", len(importances))
sns.barplot(x=features, y=importances, hue=features, palette=colors, legend=False)
plt.title("Feature Importance (RandomForestClassifier): Indicates how influential each feature is in predicting Price (target)")
plt.show()
```
